```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_data_clean = pd.read_csv('csvDatasets/Housing_Data_Clean')
```

```python
housing_data_clean.head()
```

```python
housing_data_clean.describe()
```

## House prices distribution analysis

```python
plt.figure(figsize=(8,6))
sns.histplot(housing_data_clean['price'], bins=30, kde=True)
plt.title('House prices distribution')
plt.xlabel('Price')
plt.ylabel('Amount')
plt.show()
```

## Houses area distribution

```python
plt.figure(figsize=(8,6))
sns.histplot(housing_data_clean['area'], bins=30, kde=True)
plt.title('Houses area distribution')
plt.xlabel('Area')
plt.ylabel('Amount')
plt.show()
```

## Correlation analysis

```python
plt.figure(figsize=(10,8))
sns.heatmap(housing_data_clean.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation matrix')
plt.show()
```

## Categorical data

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=housing_data_clean, x='furnishingstatus', y='price')
plt.title('Furnishing status to price')
plt.show()
```
